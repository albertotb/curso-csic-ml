Vamos a cargar un conjunto de datos de ventas de casas en King County, USA ([fuente](https://www.kaggle.com/harlfoxem/housesalesprediction))

```python
import pandas as pd
import matplotlib.pyplot as plt

house = pd.read_csv('../data/kc_house_data.csv')
```

```python
house['id'].unique().size
```

```python
house.shape
```

```python
house.drop(columns=['id', 'date', 'price']).boxplot(figsize=(8, 6))
plt.yscale('log')
plt.xticks(rotation=90);
```

```python
colnames = house.columns.difference(['id', 'date', 'price'])
ncol = 5
nrow = len(colnames) // ncol
```

```python
fig, axes = plt.subplots(nrow, ncol, figsize=(15, 8))

for idx, (ax, col) in enumerate(zip(axes.ravel(), colnames)):
    ax.scatter(house[col], house['price'])
    if idx % ncol != 0:
        ax.get_yaxis().set_visible(False)
    ax.set_title(col)
fig.tight_layout(h_pad=1)
```

Muchos modelos de aprendizaje automtico requieren que todas las variables tengan rangos similares. De lo contrario unas van a dominar las funcin objetivo que se optimiza, haciendo que el modelo no aprenda del resto. Un ejemplo son los modelos basados en distancias, como $k$ vecinos prximos. 

La clase `StandardScaler` estandariza cada columna para que tenga media 0 y desviacin 1. Todas las funciones de preproceso cuentan con la siguiente API:

   * `fit()`, "ajusta" el modelo
   * `transform()`, realiza la transformacin sobre nuevos datos, sin ajustar el modelo
   * `fit_transform()`, realiza ambas operaciones

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split

X = house.drop(columns=['id', 'date', 'price'])
y = house['price']
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled.mean(axis=0)
```

```python
X_test_scaled.mean(axis=0)
```

En este caso "ajustar" el modelo consiste en calcular la media y la desviacin de cada una de las columnas. Con `transform` usamos esas medias y desviaciones calculadas para estandarizar los datos, con la siguiente frmula:

$$\text{col_std} = \frac{\text{col} - \text{media}}{\text{desv}}$$

Es **muy importante** no usar la funcin `fit_transform` a la hora de escalar los datos de test, ya que cuando el modelo est en produccin no se va a poder calcular para nuevos datos. Por tanto, hay que hacer la transformacin con las medias y desviaciones de los datos de entrenamiento con la funcin `transform`. Podemos ver estas medias y desviaciones que ha calculado el modelo:

```python
scaler.mean_
```

```python
scaler.scale_
```

Vamos a ver la diferencia de rendimiento del modelo $k$ vecinos proximos con y sin estandarizar las variables:

```python
knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(X_train, y_train)
knn.score(X_test, y_test)
```

```python
knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(X_train_scaled, y_train)
knn.score(X_test_scaled, y_test)
```

### Otras formas de estandarizar

   * `preprocessing.MinMaxScaler`, escala los datos al intervalo $[0, 1]$
   * `preprocessing.RobustScaler`, escala los datos restando la mediana y dividiendo por el rango entre el primer y tercer cuartiles. Este escalado es ms robusto cuando hay valores extremos.
   * `preprocessing.MaxAbsScaler`, escala los datos dividiendo por el mximo valor absoluto. Este escalado no resta la media, y por tanto mantiene los la dispersin de los datos originales (nmero de 0).

### Variables categricas

Variables con un nmero de valores:

   1. discreto
   2. pequeo
   3. conocido
   
La mayora de modelos no soportan este tipo de variables directamente, ya que requieren una matriz $X$ numrica. Por tanto, siempre que esten presentes tenemos que convertirlas en columnas numricas.

```python
import seaborn as sns

diamonds = sns.load_dataset('diamonds')
diamonds.head()
```

```python
diamonds['color'].value_counts()
```

Como ya vimos, una opcin (en general mala) es transformarlas asignando un cdigo numrico a cada categora:

```python
diamonds['color_ord'] = diamonds['color'].astype('category').cat.codes
diamonds[['color', 'color_ord']].head()
```

```python
pd.Series(diamonds['color'].astype('category').cat.categories)
```

Esto es problemtico porque:

   1. introduce un orden (generalmente artificial) entre los valores
   2. introduce distancias entre los valores
   
Por ejemplo en el caso de los datos de diamantes el valor "I" est a 1 unidad de distancia de "J" pero a 4 unidades de "E". Sin embargo para otro tipo de variables como "cut", que representan niveles crecientes del corte del diamante, si podra tener sentido esta codificacin.

### Codificacin *dummy* o *one-hot*

Crea tantas variables binarias (valores 0 o 1) como el nmero de categoras. Se pueden generar con `pd.get_dummies`:

```python
color_dummy = pd.get_dummies(diamonds['color'])
```

```python
df = pd.concat((diamonds[['color', 'color_ord']], color_dummy), axis=1)
```

```python
def highlight_not_zero(s):
    not_zero = s != 0
    return ['background-color: yellow' if v else '' for v in not_zero]
    
df.head(10).style.apply(highlight_not_zero, subset=['D', 'E', 'F', 'G', 'H', 'I', 'J'])
```

Esta codificacin es redundante en el sentido de que una de las columnas queda determinada por completo a partir de las otras. Por ejemplo, la columna J solo tiene valor* uno cuando el resto de columnas tienen **todas** valor 0. Esto puede empeorar el rendimiento de modelos que tienen problemas con variables muy correladas.

La funcin `get_dummies` tiene un parmetro que elimina automticamente una de estas columnas (la primera). En general no es importante cual se elimine, pero a veces si relevante para interpretar el modelo resultante.

```python
pd.get_dummies(diamonds['color'], drop_first=True).head()
```

### Variables categricas en scikit-learn

Las dos tnicas anteriores estn implementadas en el mdulo `preprocessing` de scikit-learn:

   1. `preprocessing.OneHotEncoder`: codificacin dummy o one-hot
   2. `preprocessing.OrdinalEncoder`: codificacin ordinal

```python
from sklearn.preprocessing import OneHotEncoder

diamonds = sns.load_dataset('diamonds')

encoder = OneHotEncoder()
X = encoder.fit_transform(diamonds)
```

```python
X
```

**Ojo**: la clase `OneHotEncoder` aplica sobre **todas** las columnas. 

Para realizar una operacin sobre unas columnas y no otras tenemos que usar la clase `sklearn.compose.ColumnTransformer`, que permite aplicar distintas transformaciones a distintas columnas/grupos de columnas:

<img src=https://amueller.github.io/ml-workshop-1-of-4/slides/images/column_transformer_schematic.png width=600>

```python
diamonds['cut'].value_counts()
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder
from sklearn.compose import make_column_transformer, make_column_selector

col_trans = make_column_transformer(
    (StandardScaler(), make_column_selector(dtype_include='number')),
    (OneHotEncoder(drop='first'), ['color', 'clarity']),
    (OrdinalEncoder(categories=[['Fair', 'Good', 'Very Good', 'Premium', 'Ideal']]), ['cut'])
)
```

```python
X = col_trans.fit_transform(diamonds)
X.shape
```

```python
col_trans.transformers_
```

```python
col_trans.transformers_[2][1].categories_
```

```python
diamonds.shape
```

```python
diamonds.describe(exclude='number')
```

#### Modelos que soportan variables categricas

Tericamente, todos los modelos basados en rboles y naive Bayes soportan variables categricas sin necesidad de transformarlas a variables continuas. Sin embargo, por motivos de eficiencia no es comn implementarlos de esta forma. 

Actualmente en scikit-learn solo soportan variables categricas directamente los modelos naive Bayes.

Se prev que en prximas versiones los rboles de decisin, *random forest* y *gradient boosting* soporten variables discretas. Otras implementaciones no incluidas en scikit-learn (las veremos ms adelantes si las soportan), como por ejemplo CatBoost, si cuentan con este soporte.
